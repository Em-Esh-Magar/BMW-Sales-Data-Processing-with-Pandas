```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Transformed_Data.csv')


grouped_data = df.groupby(['Color', 'Collection_Category']).size().unstack(fill_value=0)


plt.figure(figsize=(14, 8))


bar_width = 0.15
colors = ['red', 'blue', 'green', 'orange']
x_pos = np.arange(len(grouped_data.index))


for i, category in enumerate(grouped_data.columns):
    plt.bar(x_pos + i * bar_width, grouped_data[category],
            width=bar_width, label=category, alpha=0.8)

plt.title('Car Colors Distribution Across All Collection Categories')
plt.xlabel('Color')
plt.ylabel('Number of Cars')
plt.xticks(x_pos + bar_width * (len(grouped_data.columns) - 1) / 2, grouped_data.index)
plt.legend(title='Collection Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("/content/Transformed_Data.csv")

# print(df['Model'].describe())  --> to know the unique value

grouped_data = df.groupby(['Model','Collection_Category']).size().unstack(fill_value=0)

plt.figure(figsize=(14,8))

bar_width = 0.15

x_pos = np.arange(len(grouped_data.index))

for i,category in enumerate(grouped_data.columns):
  plt.bar(x_pos + i * bar_width, grouped_data[category],width=bar_width, label=category,alpha=0.8)

plt.title('Car Models Distribution Across All Collection Categories')
plt.xlabel("Models")
plt.ylabel('Number of Cars')
plt.xticks(x_pos + bar_width * (len(grouped_data.columns) - 1) / 2, grouped_data.index)
plt.legend(title='Collection Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("/content/Transformed_Data.csv")

grouped_data = df.groupby(['Model','Price_Category']).size().unstack(fill_value=0)

plt.figure(figsize=(14,8))

bar_width = 0.25

x_pos = np.arange(len(grouped_data.index))

for i,category in enumerate(grouped_data.columns):
  plt.bar(x_pos + i * bar_width,grouped_data[category],width = bar_width, label=category, alpha=0.8)

plt.title('Car Models Distribution Across All Price Categories')
plt.xlabel("Models")
plt.ylabel('Number of Cars')
plt.xticks(x_pos + bar_width * (len(grouped_data.columns) - 1) / 2, grouped_data.index)
plt.legend(title='Collection Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
